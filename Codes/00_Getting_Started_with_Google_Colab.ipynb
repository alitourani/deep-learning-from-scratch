**A Simpe Neural Network**

```python
from numpy import loadtxt
# We are going to use Keras lib soon. An amazing library for DL
from keras.models import Sequential
from keras.layers import Dense
from tabulate import tabulate

# Loading data
print('Loading dataset ...\n')
# The dataset is Pima Indians onset of diabetes dataset
datasetUrl = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv'
dataset = loadtxt(datasetUrl, delimiter=',')
print(f'This dataset has {len(dataset)} instances, like:')

# Showing database rows
rows = ['Pregnancy', 'Plasma', 'Blood Pressure', 'Skin Fold', 'Insulin', 'BMI', 'Diabetes', 'Age', 'Output']
print(tabulate([dataset[1]], headers=rows))

# Data Splitting
print('Splitting dataset ...')
inputItems = dataset[: , 0:8]
outputItems = dataset[: , 8]
print(f'We have split the dataset into {len(inputItems)} and {len(outputItems)} items')

# Defining the model
print('Defining the model ...')
model = Sequential()
model.add(Dense(12, input_dim = 8, activation = 'relu'))
model.add(Dense(8, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))

# Compiling
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Fitting the model
print('Training the model ...')
model.fit(inputItems, outputItems, epochs=150, batch_size=10)

# Evaluating
print('Evaulation:')
_, accuracy = model.evaluate(inputItems, outputItems)
print('Accuracy: %.2f' % (accuracy*100))
```
